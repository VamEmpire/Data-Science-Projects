```python
import numpy as np
import pandas as pd
```

```python
# Loading the data and trimming it
def load_fitness_data(file_path):
  try:
    df = pd.read_csv(file_path)
    data = df[["Steps", "SleepHours", "CaloriesBurned"]].to_numpy()
    dates = df["Date"].tolist()
    return data, dates
  except Exception as e:
    print("Error loading file:", e)
    return None, None
```

- This function loads a fitness CSV file, keeps only Steps, Sleep Hours, Calories Burned, and Dates, then returns them. If an error occurs, it shows the error and returns empty values (None).

```python
# Analysing weekly data
def analyze_week(data, week_num):
  start = (week_num -1) * 7
  end = start + 7
  week = data[start:end]
  avg = np.mean(week, axis=0)
  print(f"\n Week {week_num} Avg -> Steps: {avg[0]:.0f}, Sleep: {avg[1]:.1f} hrs, Calories Burned: {avg[2]:.0f}")
  return week
```

- This function selects the required week's data, calculates the average fitness values, displays them, and returns that week's data.

```python
# Conditional Filtering
def get_low_activity_days(data, dates):
  low_steps = data[:, 0] < 6000
  poor_sleep = data[:, 1] < 6
  low_calories = data[:, 2] < 2000
  print("\nDays with low steps, sleep, or calories:")
  for i in range(len(data)):
    if low_steps[i] or poor_sleep[i] or low_calories[i]:
      print(f"{dates[i]} - Steps: {data[i, 0]}, Sleep: {data[i, 1]}, Calories: {data[i, 2]}")
```

- This function checks every day's fitness data, identifies days with low steps, poor sleep, or low calories burned, and displays those days.

```python
# Fitness Score
def fitness_score(day):
  steps, sleep, cal = day
  return 0.4 * (steps / 1000) + 0.3 * (sleep / 8) + 0.3 * (cal / 2200)
```

```python
def save_summary_to_file(data, dates, filename = "fitness_summary.txt" ):
  scores = score_all_days(data)
  best_day_idx = np.argmax(data[:, 0])
  worst_day_idx = np.argmin(data[:, 0])
  best_fitness_day = dates[np.argmax(scores)]
  worst_fitness_day = dates[np.argmin(scores)]
  top3_idx = np.argsort(scores)[-3:][::-1]

  with open(filename, "w") as f:
    f.write("Fitness Summary:\n")
    f.write(f"Average Steps: {np.mean(data[:, 0]):.0f}\n")
    f.write(f"Average Sleep: {np.mean(data[:, 1]):.1f} hrs\n")
    f.write(f"Average Calories Burned: {np.mean(data[:, 2]):.0f}\n")
    f.write(f"Best Step Day: {dates[best_day_idx]} ({data[best_day_idx, 0]:.0f} steps)\n")
    f.write(f"Worst Step Day: {dates[worst_day_idx]} ({data[worst_day_idx, 0]:.0f} steps)\n")
    f.write(f"Best Fitness Day: {best_fitness_day} (Score: {np.max(scores):.2f})\n")
    f.write(f"Worst Fitness Day: {worst_fitness_day} (Score: {np.min(scores):.2f})\n")
    f.write("Top 3 Fitness Days:\n")
    for idx in top3_idx:
      f.write(f"- {dates[idx]} (Score: {scores[idx]:.2f})\n")
  print(f"\nSummary saved to {filename}")
```
