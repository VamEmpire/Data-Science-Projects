```python
import matplotlib.pyplot as plt   
#pyplot is collection of function that make matplotlib work as MATLAB
```

```python
#1. Create first chart
```

```python
#Data
x = [1,2,3,4,5]
y = [10,20,25,35,45]

plt.plot(x,y) #line chart
plt.show() #Display chart in jupyternotebook
```

```python
#2. Customised Chart
```

```python
plt.figure(figsize=(4,3))  #Figure Size

plt.plot(x, y, color='red', marker='o', linestyle='dashed', linewidth=2, markersize=20)

plt.title("class project")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.show()
```

```python
# 3 Advanced Line Chart - multiple lines & legends
```

```python
#Data
x = [1,2,3,4,5]
y1 = [10,20,30,40,50]
y2 = [5,15,25,35,45]

plt.plot(x, y1, label='Sales 2024') #Plotting y1 Data
plt.plot(x, y2, label='sales 2025') #Plotting y2 Data

plt.title("Yearly Sales")
plt.xlabel("Months")
plt.ylabel("Sales")

plt.legend()
plt.show()
```

```python
#4. Bar Chart
```

```python
# Data
x = [1,2,3,4,5]
y = [10,15,20,25,30]

plt.xlabel("Months")
plt.ylabel("Sales")

plt.bar(x,y)
plt.title('Bar Chart1 Example')
plt.show()
```

```python
# Data
x = ['A','B','C','D','E']
y = [10,15,20,25,30]

plt.xlabel("Months")
plt.ylabel("Sales")

plt.bar(x,y)
plt.title('Bar Chart2 Example')
plt.show()
```

```python
#5. Histogram
```

```python
#Used For Distribution Analysis
# Data
import random
data = [random.randint(1, 10) for _ in range(500)]

plt.hist(data, bins=15)
plt.title("Histogram Example")
plt.show()
```

```python
#6. Pie Chart
```

```python
# Used To Show Part-to-Whole Relationship/Data

#Data
Categories = [1,2,3,4,5]
Sales = [10,15,20,25,30]

plt.pie(Sales, labels=Categories, autopct = '%.1f%%', startangle=90)  #for % percent sign
plt.title("Pie chart Example")
plt.show()
```

```python
#7. Scatter plot
# Used To Find Relationship btwn Variables
```

```python
# Data-1
Categories = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri']
Sales = [10, 20, 30, 40, 50]

#Data-2
y1 = [5,10,15,20,25]
y2 = [20,40,60,80,100]

plt.figure(figsize=(6,4))

#First Plot- Bar Chart
plt.subplot(1,2,1)  #row, column, position
plt.bar(Categories, Sales)
plt.title("weekly Sales")
plt.xlabel("week Days")
plt.ylabel("Sales")

#Second Plot- Scatter Plot
plt.subplot(1,2,2)  # Row, Column, Position
plt.scatter(y1, y2)
plt.title("Users Example")
plt.xlabel("User1")
plt.ylabel("User2")


plt.show()
```

```python
# Data-1
Categories = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri']
Sales = [10, 20, 30, 40, 50]

#Data-2
y1 = [5,10,15,20,25]
y2 = [20,40,60,80,100]

plt.figure(figsize=(10,5))

#First Plot- Bar Chart
plt.subplot(2,1,2)  #row, column, position
plt.bar(Categories, Sales)
plt.title("weekly Sales")
plt.xlabel("week Days")
plt.ylabel("Sales")

#Second Plot- Scatter Plot
plt.subplot(2,1,1)  # Row, Column, Position
plt.scatter(y1, y2)
plt.title("Users Example")
plt.xlabel("User1")
plt.ylabel("User2")


plt.show()
```

```python
#9. Matplotlib with Pandas - real data
```

```python
#Create Dataframe
import pandas as pd

#Data
data = {
    'Month' : ['Jan', 'Feb', 'March', 'April'],
    'Sales' : [12000, 11000, 13000, 25000]
}

df = pd.DataFrame(data)
```

```python
df
```

```python
plt.bar(df['Month'],df['Sales'])  #x  and #y
plt.title("Matplotlib with pandas")
plt.xlabel("Month")
plt.ylabel("Sales")

plt.show()
```

```python
#10. Save Plot/Charts
```

```python
plt.bar(df['Month'], df['Sales'])
plt.title("Matplotlib with Pandas")
plt.xlabel("Month")
plt.ylabel("Sales")

plt.savefig("Monthly_Sales_from_df.png")

plt.show()
```
