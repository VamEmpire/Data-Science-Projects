## 3. Array initialization methods

```python
# zeros array

zero_arr = np.zeros((2,3))
print(zero_arr)
```

```python
# zeros array
zero_arr = np.zeros((2,3),dtype = int) # HyperParameter Tuning
print(zero_arr)
```

```python
# Ones array
ones_arr = np.ones((2,3))
print(ones_arr)
```

```python
# Full array

full_arr = np.full((3,2), 9)
print(full_arr)
```

```python
# Identity Matrix

id_arr = np.eye(3)
print(id_arr)
```

```python
# Empty array
print(np.empty(2))
```

```python
# Evenly spaced array
np.arange(3.0)
```

```python
# Evenly spaced array
print(np.arange(1,10,2))
```

```python
# Evenly spaced array
print(np.arange(1,3,1))
```

```python
# Evenly spaced array
print(np.arange(2,10,2))
```

```python
# Specific values between a range
print(np.linspace(1,10,4))
```

```python
# random values array
r_arr = np.random.rand(3,2)
print(r_arr)
```

```python
# random values 
r_arr = np.random.randint(1,10,(3,3))
print(r_arr)
```
